```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("clean_crash_data.csv")
data
```

```python
mask = data.injury_severity != "NO APPARENT INJURY"
column_name = "injury_severity"
data.loc[mask, column_name] = "INJURY"
data['unique_id'] = range(1, len(data) + 1)
```

```python
data.groupby("vehicle_make")[["unique_id"]].nunique().\
sort_values(by=["unique_id"], ascending=False).head(50)
```

```python
data.groupby("injury_severity")[["unique_id"]].nunique()
```

```python
pd.get_dummies(data[["injury_severity"]])[["injury_severity_INJURY"]]
data.columns
```

```python
y = pd.get_dummies(data[["injury_severity"]])[["injury_severity_INJURY"]]
X = pd.get_dummies(data[["vehicle_make", "vehicle_dmg_extent", "weather"]])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import balanced_accuracy_score as bas
from sklearn.metrics import precision_score as ps
from sklearn.metrics import mean_squared_error as mse
classifier.score(X_test, y_test)
print(f"accuracy: {classifier.score(X_test, y_test)}")
print(f"balanced acc: {bas(classifier.predict(X_test), y_test)}")
print(f"mean sq error: {mse(classifier.predict(X_test), y_test)}")
print(f"precision: {ps(classifier.predict(X_test), y_test)}")
```
