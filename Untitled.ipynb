```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
crash_data_raw = pd.read_csv("crash_data.csv").\
drop(columns= ["Off-Road Description", "Municipality", "Related Non-Motorist"])

time_date_data = crash_data_raw["Crash Date/Time"].str.split()
names = ["date", "time", "AM/PM"]
time_date = pd.DataFrame(time_date_data.to_list(), columns=names)
time_date["time"] = time_date["time"] + time_date["AM/PM"]
time_date = time_date.drop(columns="AM/PM")
crash_data_raw["time"] = time_date["time"]
crash_data_raw["date"] = time_date["date"]

crash_data_raw
```

```python
pd.get_dummies(crash_data_raw[["Injury Severity"]])[["Injury Severity_FATAL INJURY"]]
```

```python
crash_data_raw.groupby("Vehicle Make")[["Report Number"]].nunique().\
sort_values(by=["Report Number"], ascending=False).head(50)
```

```python
y = crash_data_raw[["Driver At Fault"]]
X = crash_data_raw[["Vehicle Year"]]
y

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
df = pd.DataFrame({"P": classifier.predict(X_test)})
df["A"] = np.array(y_test)

df.head(50)
classifier.score(X_test, y_test)
classifier.predict()
```
